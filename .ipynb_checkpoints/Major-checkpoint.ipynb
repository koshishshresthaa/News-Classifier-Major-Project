```python
nltk.download('stopwords')
nepali_stopwords=stopwords.words('nepali')
print(nepali_stopwords)
```

```python
# adding punctuation file
```

```python
punctuation_file = 'D:/Major Project/datas/nepali_punctuation.txt'
punctuation_words = []
with open(punctuation_file, encoding='utf-8') as fp:
    lines = fp.readlines()
    punctuation_words =list( map(lambda x:x.strip(), lines))
punctuation_words
```

```python
#removing stop words , punctuations and noises
```

```python
from snowballstemmer import NepaliStemmer

def preprocess_text(cat_data, nepali_stopwords, punctuation_words):
    stemmer = NepaliStemmer() # initialize the Nepali stemmer
  
    new_cat = []
    noise = "1,2,3,4,5,6,7,8,9,0,०,१,२,३,४,५,६,७,८,९".split(",")
  
    for row in cat_data:
        words = row.strip().split(" ")      
        nwords = ""
    
        for word in words:
            # apply Nepali stemming to the word
            if word not in punctuation_words and word not in nepali_stopwords:
                word = stemmer.stemWord(word)
                
                is_noise = False
                for n in noise:
                    if n in word:
                        is_noise = True
                        break
                if not is_noise and len(word) > 1:
                    word = word.replace("(","")
                    word = word.replace(")","")
                    nwords += word + " "
                
          
        new_cat.append(nwords.strip())
  
    return new_cat
```

```python
title_clean = preprocess_text(["\n मेरो नाम कोशिश श्रेष्ठ हो ।  म काठमाडौँमा बस्छु । म राम्रो मान्छे बन्न चाहन्छु । "], nepali_stopwords,punctuation_words)
print(title_clean)
```
